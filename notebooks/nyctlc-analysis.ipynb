```python
# hourly dropoff and pickup trip counts for individual zones
# db_path = 'data/processed/nyctlc-2012.db'
db_path = 'E:/Work/projects/twitterinfrastructure/data/processed/nyctlc-2012.db'

datetime_range_ref = ('2012-01-01 00:00:00', '2012-12-31 23:59:59')
datetime_range_excl = ('2012-10-22 00:00:00', '2012-11-02 23:59:59')

# tuple = (dayofweek, hour, zone)
# dayofweek (0 = Monday)
# hour (4 = 04:00, 16 = 16:00)
# zones:
# 233 = UN/Turtle Bay South,
# 74 = East Harlem North, 
# 186 = Penn Station/Madison Sq West, 
# 33 = Brooklyn Heights, 
# 230 = Times Sq/Theatre District

combos = [(6, 16, 230), (3, 16, 230), 
          (6, 16, 186), (3, 16, 186)]

# query reference data
sql = """
    SELECT dropoff_datetime AS datetime, dropoff_location_id AS zone,
        mean_pace, trip_count 
    FROM summary_zonedropoff_hour_2012
    WHERE
        (dropoff_datetime BETWEEN "{start_datetime}" AND "{end_datetime}")
        AND (dropoff_datetime NOT BETWEEN "{start_datetime_excl}" AND
            "{end_datetime_excl}")
;""".format(start_datetime=datetime_range_ref[0],
            end_datetime=datetime_range_ref[1],
            start_datetime_excl=datetime_range_excl[0],
            end_datetime_excl=datetime_range_excl[1])
df = query(db_path, sql)
df['mean_pace'] = df['mean_pace'] / 60  # convert mean_pace to [min./mile]

# add dayofweek (0 = Monday) and hour (0-23)
df['datetime'] = pd.to_datetime(df['datetime'])
df['dayofweek'] = df['datetime'].dt.dayofweek
df['hour'] = df['datetime'].dt.hour

# filter to desired dayofweek, hour, and zone combos
df_plot = pd.DataFrame()
for combo in combos:
    df_plot = df_plot.append(df[(df['dayofweek'] == combo[0]) &
                                (df['hour'] == combo[1]) &
                                (df['zone'] == combo[2])])
df_plot['timegroup'] = [str(day) + '-' + str(hour) for day, hour in 
                        zip(df_plot['dayofweek'].values, 
                            df_plot['hour'].values)]
df_plot = df_plot[['zone', 'timegroup', 'trip_count']]
df_plot = df_plot.rename(columns={'trip_count': 'trip-count'})
df_plot.head()

# dropoff ztrips boxplot
fig, ax = plt.subplots(figsize=(6, 4), tight_layout=False)
# ax.boxplot([ploty1, ploty2, ploty3, ploty4, ploty5, ploty6], sym='', 
#            labels=['W-16:00-233', 'Sun.-04:00-233',
#                    'W-16:00-74', 'Sun.-04:00-74',
#                    'W-16:00-186', 'Sun.-04:00-186'])
ax = sns.boxplot(x='zone', y='trip-count', hue='timegroup', data=df_plot,
                 ax=ax)
ax.set_xlabel('Zone', fontsize=16)
ax.set_ylabel('\# of Dropoffs', fontsize=16)
ax.tick_params(axis='both', labelsize=14)
labels = ['Sunday, 16:00', 'Thursday, 16:00']
handles, old_labels = ax.get_legend_handles_labels()
leg = ax.legend(handles=handles, labels=labels, loc='best', fontsize=12, 
                title=None, frameon=True, fancybox=True, 
                facecolor='xkcd:light grey')
print('old labels: ' + str(old_labels))

# save
# plt.savefig('notebooks/images/dropoffs-boxplot.png', dpi=150, bbox_inches='tight')
# plt.savefig('./images/dropoffs-boxplot.png', dpi=150, bbox_inches='tight')

# 
# dropoff ztrips histogram
plot1 = df[(df['dayofweek'] == combos[0][0]) &
           (df['hour'] == combos[0][1]) &
           (df['zone'] == combos[0][2])]['trip_count']
fig, ax = plt.subplots(figsi ze=(6, 4), tight_layout=False)
ax.hist(plot1, density=True, facecolor='g', alpha=0.75)
ax.set_xlabel('Dropoffs (TR, 16:00, Zone 230)', fontsize=16)
ax.set_ylabel('Probability', fontsize=16)
ax.tick_params(axis='both', labelsize=14)

# save
# plt.savefig('notebooks/images/dropoffs-hist.png', dpi=150, bbox_inches='tight')
# plt.savefig('./images/dropoffs-hist.png', dpi=150, bbox_inches='tight')

```

```python
# hourly zone dropoff zscores
# db_path = 'data/processed/nyctlc-2012.db'
db_path = 'E:/Work/projects/twitterinfrastructure/data/processed/nyctlc-2012.db'

trip_count_filter = 5   # by hour
startdate_sql = '2012-10-28 00:00:00'  # inclusive
enddate_sql = '2012-11-02 23:59:59'  # inclusive
sql = """
        SELECT dropoff_datetime AS datetime, dropoff_location_id AS zone, 
            z_mean_pace AS zpace_drop, z_trip_count AS ztrips_drop 
        FROM standard_zonedropoff_hour_sandy
        WHERE 
            trip_count > {trip_count_filter} AND
            dropoff_datetime BETWEEN "{startdate_sql}" AND "{enddate_sql}"
      ;""".format(trip_count_filter=trip_count_filter,
                  startdate_sql=startdate_sql, enddate_sql=enddate_sql)
df_drop = query(db_path, sql)
df_drop['datetime'] = pd.to_datetime(df_drop['datetime'])
df_drop = df_drop.rename(columns={'zpace_drop': 'zpace-drop',
                                  'ztrips_drop': 'ztrips-drop'})
df_drop['abs-zpace-drop'] = abs(df_drop['zpace-drop'])
df_drop['abs-ztrips-drop'] = abs(df_drop['ztrips-drop'])
df_drop = df_drop[['datetime', 'zone', 'abs-zpace-drop', 'zpace-drop', 
                   'abs-ztrips-drop', 'ztrips-drop']]
print('[min, max] dropoff datetime: [' + 
      str(np.nanmin(df_drop['datetime'])) + ', ' + 
      str(np.nanmax(df_drop['datetime'])) + '].')
print('[min, max] dropoff pace and trips z-score: [' + 
      str(np.nanmin(df_drop['zpace-drop'])) + ', ' +
      str(np.nanmax(df_drop['zpace-drop'])) + '], [' +
      str(np.nanmin(df_drop['ztrips-drop'])) + ', ' +
      str(np.nanmax(df_drop['ztrips-drop'])) + '].')
print('')

ploty = df_drop['ztrips-drop']

# dropoff ztrips boxplot
fig, ax = plt.subplots(figsize=(6, 4), tight_layout=False)
ax.boxplot(ploty, sym='', labels=['Oct. 28 - Nov. 2'])
ax.set_ylabel(r'$z_d$', fontsize=16)
ax.tick_params(axis='both', labelsize=16)

# dropoff ztrips histogram
fig, ax = plt.subplots(figsize=(6, 4), tight_layout=False)
# ax.hist(ploty, density=True, facecolor='g', alpha=0.75)
ax.hist(ploty, np.arange(-2, 5, 0.25), density=False, facecolor='g', alpha=0.75)
ax.set_xlabel(r'$z_d$ (Oct. 28 - Nov. 2)', fontsize=16)
ax.set_ylabel('Count', fontsize=16)
```
